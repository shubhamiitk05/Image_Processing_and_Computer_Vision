```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
# FOR IMAGE 1

image = cv2.imread('cctv1.JPG')

# Obtaining the breadth and length of the image
breadth = image.shape[1]
length = image.shape[0]

# Converting the image to grayscale
gray = cv2.cvtColor(image, code=cv2.COLOR_BGR2GRAY)

# Increase contrast of the image
contrast_image = cv2.multiply(gray, 2)

# Adjust the brightness of the image
brightness_image = np.uint8(np.power(contrast_image / 255, 0.5) * 255)

# Clip the pixel values to ensure they are within the valid range
new_image = np.clip(brightness_image, 0, 255)


# Displaying the enhanced image
cv2_imshow( new_image)

# Saving the enhanced image as a JPEG file
cv2.imwrite("./enhanced-cctv1.jpg", new_image)
```

```python
# FOR IMAGE 2

image = cv2.imread('cctv2.JPG')

# Obtaining the breadth and length of the image
breadth = image.shape[1]
length = image.shape[0]

# Converting the image to grayscale
gray = cv2.cvtColor(image, code=cv2.COLOR_BGR2GRAY)
# Increase contrast of the image
contrast_image = cv2.multiply(gray, 1/3)

# Adjust the brightness of the image
brightness_image = np.uint8(np.power(contrast_image / 255, 1/3) * 255)

# Clip the pixel values to ensure they are within the valid range
new_image = np.clip(brightness_image, 0, 255)

# Displaying the enhanced image
cv2_imshow( new_image)

# Saving the enhanced image as a JPEG file
cv2.imwrite("./enhanced-cctv2.jpg", new_image)
```
