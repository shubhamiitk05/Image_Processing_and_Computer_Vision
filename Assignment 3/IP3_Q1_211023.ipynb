```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# FOR IMAGE 1
image=cv2.imread(f'gutters1.JPG')

# Converting the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Applying adaptive thresholding
block_size = 9
threshold_value = 11

# Using adaptive thresholding to create a binary image
binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, threshold_value)

# Displaying the output image
cv2_imshow(binary)
```
